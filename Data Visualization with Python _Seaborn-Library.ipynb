```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# SeaBorn Line Plot
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.show()
```

```python
fmri.shape
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event')
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event',markers=True)
plt.show()
```

```python
# SeaBorn Bar Plot
```

```python
import pandas as pd
sns.set(style='whitegrid')
pokemon=pd.read_csv('pokemon.csv')
```

```python
sns.barplot(x='is_legendary',y='speed',data=pokemon)
plt.show()
```

```python
sns.barplot(x='is_legendary',y='weight_kg',data=pokemon,palette='vlag') # palette='Blue_d','rocket','vlag'
plt.show()
```

```python
sns.barplot(x='is_legendary',y='speed',hue='generation',data=pokemon)
plt.show()
```

```python
sns.barplot(x='is_legendary',y='speed',data=pokemon,color='orange')
plt.show()
```

```python
# SeaBorn Hist Plot (Distribution and Histogram Plot)
```

```python
diamonds = pd.read_csv('diamonds.csv')
diamonds.head()
```

```python
sns.distplot(diamonds['price'])
plt.show()
```

```python
sns.distplot(diamonds['price'],hist=False)
plt.show()
```

```python
sns.distplot(diamonds['price'],color='purple')
plt.show()
```

```python
sns.distplot(diamonds['price'],bins=15,kde=False)
plt.show()
```

```python
sns.distplot(diamonds['price'],color='orange',vertical=True)
plt.show()
```

```python
# SeaBorn JointPlot (Scatterplot and HistogramPlot)
```

```python
iris = pd.read_csv('iris_csv.csv')
sns.set(style='white')
sns.jointplot(x='sepallength',y='petallength',data=iris)
plt.show()
```

```python
sns.jointplot(x='sepallength',y='petallength',data=iris,color='olive')
plt.show()
```

```python
sns.jointplot(x='sepallength',y='petallength',data=iris,kind='reg')
plt.show()
```

```python
# SeaBorn BoxPlot
```

```python
churn = pd.read_csv('churn.csv')
churn.head()
```

```python
sns.boxplot(x='Churn',y='tenure',data=churn)
plt.show()
```

```python
sns.boxplot(x='InternetService',y='MonthlyCharges',data=churn)
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,palette='Set1')
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,color='orange',linewidth=3)
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,order=['Two year','Month-to-month','One year'],linewidth=3)
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,hue='PaymentMethod')
plt.show()
```
